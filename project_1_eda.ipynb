#### *Explanation of the columns in the given index*:

1. **App Name**: The name or title of the mobile application.
2. **App Id**: A unique identifier for the application within the dataset.
3. **Category**: The category or genre to which the application belongs.
4. **Rating**: The average user rating or evaluation of the application.
5. **Rating Count**: The total count of user ratings received by the application.
6. **Installs**: The number of times the application has been installed.
7. **Minimum Installs**: The minimum number of installations required for the application.
8. **Maximum Installs**: The maximum number of installations reached by the application.
9. **Free**: Indicates whether the application is available for free or not (Boolean: True or False).
10. **Price**: The cost of the application if it is not available for free.
11. **Currency**: The currency in which the price of the application is specified.
12. **Size**: The size of the application in terms of storage space.
13. **Minimum Android**: The minimum Android version required to run the application.
14. **Developer Id**: Unique identifier for the developer of the application.
15. **Developer Website**: The website associated with the developer of the application.
16. **Developer Email**: The email address of the developer.
17. **Released**: The date when the application was initially released.
18. **Last Updated**: The most recent date when the application was updated.
19. **Content Rating**: The age group or audience for which the application is suitable.
20. **Privacy Policy**: Information about the application's privacy policy.
21. **Ad Supported**: Indicates whether the application includes advertisements (Boolean: True or False).
22. **In App Purchases**: Indicates whether the application offers in-app purchases (Boolean: True or False).
23. **Editors Choice**: Indicates whether the application has been selected as an editor's choice (Boolean: True or False).
24. **Scraped Time**: The timestamp when the data was collected or scraped.

These columns provide various attributes and information about the mobile applications within the dataset, allowing for analysis and insights into different aspects of app characteristics and user interactions.

```python
### Data Cleaning
# # List of characters to remove
chars_to_remove = ['+', ',', '$', 'M', 'k']
# List of column names to clean
cols_to_clean = ['Installs','Size', 'Price']

# Loop for each column
for i in cols_to_clean:
    if apps[i].dtype == 'object':  # check if column is of string type
        for char in chars_to_remove:
            apps[i] = apps[i].str.replace(char, '')
        apps[i] = pd.to_numeric(apps[i], errors='coerce')  # convert column to numeric
```

```python
# Which apps were free  in each category?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\64284\OneDrive\Desktop\ai_with_ammar\day_8\google-playstore.csv")

# Count the number of free and paid apps in each category
category_counts = df.groupby(['Category', 'Free']).size().unstack(fill_value=0)

# Plot a stacked bar chart of the category counts
category_counts.plot(kind='bar', stacked=True)

# Set the labels and title for the plot
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Free and Paid Apps in Category")

# Display the plot
plt.show()
```

```python
# Correlation between features
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
df = pd.read_csv(r'C:\Users\64284\OneDrive\Desktop\ai_with_ammar\day_8\google-playstore.csv')

# Select the columns to include in the heatmap
cols = ['Rating', 'Rating Count', 'Minimum Installs', 'Maximum Installs', 'Price']

# Convert the selected columns to a numeric data type
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

# Calculate the correlation between the columns
corr = df[cols].corr()

# Create the heatmap using seaborn
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Set the title of the heatmap
plt.title('Correlation Heatmap')

# Display the heatmap
plt.show()
```
